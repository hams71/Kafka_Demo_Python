```python

from kafka import KafkaConsumer
import json
from kafka.errors import KafkaError
import requests
import time
from pyspark.sql import SparkSession
from pyspark.sql.functions import count
```

```python
REST_API = "<your-api>

def json_deserializer(data):
    return json.loads(data)

# each consumer will be in a consumer group
consumer = KafkaConsumer(
                        bootstrap_servers=["localhost:9092"],
                        auto_offset_reset='earliest',
                        group_id="C1",
                        key_deserializer=json_deserializer,
                        value_deserializer=json_deserializer)
consumer.subscribe(["Reg_User"])


dic = {}

if __name__ == "__main__":
    print("Starting")
    for msg in consumer:

        data = msg.value

        temp = data["country"]
        #dic[data["country"]] = 0
        if temp not in dic.keys():
             dic[data["country"]] = 0
             dic[data["country"]] += data["price"]
        else:
            dic[data["country"]] += data["price"]
            print(dic.keys())
            print(dic.values())
   
        
        msg.value["price"] = dic[msg.value["country"]]
        
        
        print(data["price"])
        print(type(data))

        #print(msg)
        print(msg.value)
        headers = {'content-type': 'application/json'}
        # have to the have the json in list [values] for it to work
        r = requests.post(url=REST_API,data=json.dumps([msg.value]),headers=headers)
        print(r.status_code)
        time.sleep(2)
        
```
